```python
# load data
week1_data = pd.read_csv('data/week1.csv')
week2_data = pd.read_csv('data/week2.csv')

display(week1_data.head(1))
display(week2_data.head(2))
```

After loading in data from weeks 1 and 2 we can see that the columns have incorrect names and that the correct names are in rows 0 and 1 respectively. This suggests that the data has to be loaded and processed case by case due to formatting differences.

Loading in the data with correct columns using skiprows reveals an empty column (Unnamed: 0), missing values along with features unrelated to the price of turnips such as Contact(Twitter/Discord), In-Game Name, Island and etc.

```python
# load data with correct column names
week1_data = pd.read_csv('data/week1.csv', skiprows=[0])
week2_data = pd.read_csv('data/week2.csv', skiprows=[0, 1])

display(week1_data.head(3))
display(week2_data.head(3))

print(week1_data.dtypes)
```

Based on the data samples shown above the turnip prices appear to be fine apart from missing values while further inspection reveals an issue related to the object data type which indicates the presence of mixed data where we would expect to only see floats.
